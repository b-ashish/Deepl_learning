```python
import numpy as np
import pandas as pd

from glob import glob
from sklearn.model_selection import train_test_split

from tensorflow.keras.layers import Input,Dense,Flatten
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
```

```python
train_path = (r'C:\Users\ASUS\Documents\DS_class\model pract\DL\crop_disease\project_App\data\Cotton plant')
test_path = (r'C:\Users\ASUS\Documents\DS_class\model pract\DL\crop_disease\project_App\data\Cotton plant disease')
```

```python
img_size = [512,512]
```

```python
incpt = InceptionV3(input_shape= img_size+[3],weights='imagenet',include_top=False)
```

```python
for layers in incpt.layers:
    layers.trainable=False
```

```python
folder = glob(r"C:\Users\ASUS\Documents\DS_class\model pract\DL\crop_disease\project_App\data\Cotton plant/*")
```

```python
x=Flatten()(incpt.output)
```

```python
prediction = Dense(len(folder),activation = 'softmax')(x)
```

```python
model = Model(incpt.input,outputs=prediction)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale =1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_set = train_datagen.flow_from_directory(train_path,target_size=(512,512),batch_size=32,class_mode='categorical')
test_set =  test_datagen.flow_from_directory(test_path,target_size=(512,512),batch_size=32,class_mode='categorical')
```

```python
classes = train_set.class_indices
classes
```

```python
fit = model.fit(train_set,validation_data=test_set,epochs=10,steps_per_epoch=len(train_set),
                         validation_steps=len(test_set))
```

```python
pred_img_path = (r"C:\Users\ASUS\Documents\DS_class\model pract\DL\crop_disease\project_App\ctn_crp.jpg")
pred_img_preprocess = image.load_img(pred_img_path,target_size=(512,512))
inp_array = image.img_to_array(pred_img_preprocess)
```

```python
img_arr = np.array([inp_array])
img_arr = img_arr.astype('float32')/255
```

```python
pred = model.predict(img_arr)
pred_class = np.argmax(pred,axis=1)
pred_class
```

```python
for i,j in enumerate(classes):
    if i==pred_class[0]:
        print(f"Predicted Cotton shows {j} nature")
```
